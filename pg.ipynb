```python
import pandas as pd
import json
```

```python
def loadFundInfo(csvpath):
    df = pd.read_csv(csvpath)
    fundname = json.loads(df['date'][0].replace('""', '"'))['name']
    df = df[1:][["date", "netvalue"]]
    df['date'] = pd.to_datetime(df['date'])
    df.set_index('date', inplace=True)
    df.rename(columns={'netvalue': fundname}, inplace=True)
    df.columns.name = "Ticker"
    return df
```
